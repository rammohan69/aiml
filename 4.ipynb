```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
df = pd.read_csv("C:\\Users\\manje\\OneDrive\\Desktop\\pradhanmatri\\breast_cancer.csv")
print(df.shape)
df.head()
```

```python
X=df.iloc[:,2:].values
Y=df.iloc[:,1].values
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=100)
```

```python
model=GaussianNB()
model.fit(X_train,Y_train)
```

```python
train_accuary=accuracy_score(Y_train,model.predict(X_train))
test_accuary=accuracy_score(Y_test,model.predict(X_test))
print(train_accuary)
print(test_accuary)
```

```python
train_confusion=confusion_matrix(Y_train,model.predict(X_train))
test_confusion=confusion_matrix(Y_test,model.predict(X_test))
print(train_confusion)
print(test_confusion)
```

```python
train_classifier=classification_report(Y_train,model.predict(X_train))
test_clasifier=classification_report(Y_test,model.predict(X_test))
print(train_classifier)
print(test_clasifier)
```

```python
begnin=test_confusion[0,0]+test_confusion[0,1]
malignent=test_confusion[1,0]+test_confusion[1,1]
print(begnin)
print(malignent)
```
