```python
varities=list(seed_df.pop('grain_variety'))
```

```python
samples=seed_df.values
```

```python
from scipy.cluster.hierarchy import linkage,dendrogram
```

```python
mergings=linkage(samples,method='complete')
```

```python
import matplotlib.pyplot as plt
dendrogram(mergings,labels=varities,leaf_rotation=90,leaf_font_size=10)
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster
```

```python
labels=fcluster(mergings,1,criterion='distance')
```

```python
df=pd.DataFrame({'labels':labels,'varities':varities})
```

```python
ct=pd.crosstab(df['labels'],df['varities'])
ct
```
