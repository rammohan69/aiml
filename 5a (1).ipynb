```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dp=pd.read_csv("C:\\Users\\manje\\OneDrive\\Desktop\\pradhanmatri\\ch1ex1.csv")
print(dp.shape)
dp.head()
```

```python
points=dp.values
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
```

```python
model.fit(points)
```

```python
lable=model.predict(points)
x=points[:,0]
y=points[:,1]
plt.scatter(x,y,c=lable)
```

```python
centroid=model.cluster_centers_
```

```python
centroid_x=centroid[:,0]
centroid_y=centroid[:,1]
```

```python
plt.scatter(x,y,c=lable)
plt.scatter(centroid_x,centroid_y,marker='x',s=2000)
plt.show()
```
